Задача 1: Фильтрация списка (Lists + For)
Напиши функцию filter_and_double(numbers), которая принимает список чисел. Функция должна:
Выбрать только четные числа.
Каждое из этих чисел умножить на 2.
Вернуть результат в виде нового списка.

```python
def filter_and_double(numbers):
  even_numbers_x2 = [i*2 for i in numbers if i % 2 == 0]
  return even_numbers_x2

print (filter_and_double([1, 2, 3, 4, 5, 6]))
print (filter_and_double([10, 11, 13, 20]))
print (filter_and_double([1, 3, 5]))

assert filter_and_double([1, 2, 3, 4, 5, 6]) == [4, 8, 12]
assert filter_and_double([10, 11, 13, 20]) == [20, 40]
assert filter_and_double([1, 3, 5]) == []
```

Задача 2: Магазин (Dictionaries + For)
У тебя есть словарь prices, где ключ — название товара, а значение — его цена. И есть список покупок cart. Напиши функцию calculate_total(cart, prices), которая посчитает итоговую стоимость корзины.
Если товара из корзины нет в словаре цен, просто игнорируй его.

```python
store_prices = {"яблоко": 50, "банан": 30, "хлеб": 20, "молоко": 100}

def calculate_total(cart, prices):
    total = 0
    for i in cart:
        if i in prices:
            total += prices[i]
    return total

print (calculate_total(["яблоко", "банан"], store_prices))
print (calculate_total(["хлеб", "хлеб", "молоко"], store_prices))
print (calculate_total(["печенье", "яблоко"], store_prices))

assert calculate_total(["яблоко", "банан"], store_prices) == 80
assert calculate_total(["хлеб", "хлеб", "молоко"], store_prices) == 140
assert calculate_total(["печенье", "яблоко"], store_prices) == 50
```

Задача 3: Поиск в кортеже (Tuples + While)
Напиши функцию find_first_index(data, target), которая использует цикл while. Она должна найти индекс первого вхождения элемента target в кортеже data. Если элемент не найден, функция должна вернуть -1.

```python
my_tuple = (10, 20, 30, 40, 20, 50)

def find_first_index(data, target):
    i = 0
    while i < len(data):
        if data[i] == target:
            return i
        i += 1
    return -1

print (find_first_index(my_tuple, 20))
print (find_first_index(my_tuple, 50))
print (find_first_index(my_tuple, 100))

assert find_first_index(my_tuple, 20) == 1
assert find_first_index(my_tuple, 50) == 5
assert find_first_index(my_tuple, 100) == -1
```

Задача 4: Статистика символов (Dicts + For + Strings)
Напиши функцию count_letters(text), которая принимает строку и возвращает словарь. Ключами должны быть буквы, а значениями — то, сколько раз они встретились в строке. Пробелы учитывать не нужно.


```python
def count_letters(text):
  result = {}
  for i in text:
    if i != " ":
      if i in result:
        result[i] += 1
      else:
        result[i] = 1
  return result

print (count_letters("hello"))

assert count_letters("hello") == {"h": 1, "e": 1, "l": 2, "o": 1}
assert count_letters("aba") == {"a": 2, "b": 1}
assert count_letters("") == {}
```

Задача 6: Система регистрации (Sets + Dicts + Functions)
В этой задаче нужно реализовать две функции для управления курсами студентов. Мы будем использовать множества, так как курсы не должны дублироваться.

add_courses(db, name, *courses) — добавляет студенту name список курсов. Если студент уже есть, новые курсы добавляются к старым.

get_common_courses(db, name1, name2) — возвращает множество курсов, которые посещают оба студента.

```python
def add_courses(db, name, *courses):
    if name not in db:
        db[name] = set()
    db[name].update(courses)

def get_common_courses(db, name1, name2):
    if name1 not in db or name2 not in db:
        return set()
    return db[name1] & db[name2]

students_db = {}

add_courses(students_db, "Аня", "Python", "SQL")
add_courses(students_db, "Аня", "Алгоритмы", "Python")
add_courses(students_db, "Иван", "Python", "Django")

print (students_db)
print (get_common_courses(students_db, "Аня", "Иван"))
print (get_common_courses(students_db, "Аня", "Борис"))

assert students_db["Аня"] == {"Python", "SQL", "Алгоритмы"}
assert get_common_courses(students_db, "Аня", "Иван") == {"Python"}
assert get_common_courses(students_db, "Аня", "Борис") == set()
```

9 Напиши функцию process_data(items, *, unique=False, threshold=None, sort_reverse=False).

Логика работы:

unique: если True, функция должна оставить только уникальные элементы (используй set).

threshold: если передано число, нужно оставить только те элементы, которые строго больше этого значения.

sort_reverse: если True, итоговый список должен быть отсортирован по убыванию. Если False — по возрастанию.

Важно: Обрати внимание на * в аргументах. Это значит, что параметры unique, threshold и sort_reverse можно передать только по имени (например, unique=True), но не просто по порядку.

```python
def f(a, b, *, c=None, d=None):
  print(a, b, c, d)
f(1, 2, c="x")
```

```python
numbers = [10, 2, 5, 10, 1, 8, 2]

def process_data(items, *, unique=False, threshold=None, sort_reverse=False):
    result = list(items)

    if unique:
        result = list(set(result))

    if isinstance(threshold, (int, float)):
        filtered_result = []
        for i in result:
          if i > threshold:
            filtered_result.append(i)
        result = filtered_result

    if sort_reverse is True:
        result.sort(reverse=True)
    else:
        result.sort()

    return result


assert process_data(numbers) == [1, 2, 2, 5, 8, 10, 10]
assert process_data(numbers, unique=True, sort_reverse=True) == [10, 8, 5, 2, 1]
assert process_data(numbers, unique=True, threshold=5) == [8, 10]
assert process_data(numbers, threshold=100) == []
assert process_data([], unique=True) == []

print (process_data(numbers))
print (process_data(numbers, unique=True, sort_reverse=True))
print (process_data(numbers, threshold=100))
print (process_data(numbers))
print (process_data([], unique=True))
```

10 Напиши функцию trim_borders(data, k), которая удаляет k элементов с начала и k элементов с конца списка.

Если k слишком велико и элементов не остается — верни пустой список.

Если k равно 0 — верни копию списка.


```python
def trim_borders(data, k):
  if k >= len(data):
    return []
  elif k == 0:
    return data
  else:
    return data[k:-k]

assert trim_borders([1, 2, 3, 4, 5], 1) == [2, 3, 4]
assert trim_borders([1, 2, 3, 4, 5], 2) == [3]
assert trim_borders([1, 2], 5) == []
assert trim_borders([10, 20], 0) == [10, 20]

print(trim_borders([1, 2, 3, 4, 5], 1))
print(trim_borders([1, 2, 3, 4, 5], 2))
print(trim_borders([1, 2], 5))
print(trim_borders([10, 20], 0))
```

Напиши функцию deep_flatten(lst), которая принимает список с любым уровнем вложенности и возвращает один плоский список.

Подсказка: ключевые слова тут это рекурсия и isinstance     

Пройдись циклом по data:
Если это список -> рекурсивно добавь его элементы в result
Если это число/строка -> просто добавь в result

```python
def deep_flatten(data):
  result = []
  for i in data:
    if isinstance(i, list):
      result.extend(deep_flatten(i)) # почему i а не data
    else:
      result.append(i)
  return result

print (deep_flatten([1, [2, [3, 4], 5], 6]))
print (deep_flatten([[[]]]))
print (deep_flatten([1, 2, 3]))
print (deep_flatten(["a", ["b", ["c"]]]))

assert deep_flatten([1, [2, [3, 4], 5], 6]) == [1, 2, 3, 4, 5, 6]
assert deep_flatten([[[]]]) == []
assert deep_flatten([1, 2, 3]) == [1, 2, 3]
assert deep_flatten(["a", ["b", ["c"]]]) == ["a", "b", "c"]
```

```python
def deep_sum(lst):
  numbers = 0
  for i in lst:
    if isinstance(i, list):
      numbers += deep_sum(i)
    else:
      numbers += i
  return numbers

print (deep_sum([1, [2, [3, 4]]]))
print (deep_sum([[10, 20], [30, 40]]))
print (deep_sum([[[[]]]]))

# Проверки:
assert deep_sum([1, [2, [3, 4]]]) == 10
assert deep_sum([[10, 20], [30, 40]]) == 100
assert deep_sum([[[[]]]]) == 0
```

У тебя есть список сотрудников, где каждый сотрудник — это словарь. Напиши функцию group_by_department(employees), которая превратит этот список в словарь, где ключи — названия отделов, а значения — списки имен сотрудников этого отдела.


Подсказка: создавай пустой список в словаре, если такого отдела еще не встречалось.

```python
staff = [
    {"name": "Алексей", "dept": "IT"},
    {"name": "Мария", "dept": "HR"},
    {"name": "Иван", "dept": "IT"},
    {"name": "Елена", "dept": "Design"},
    {"name": "Олег", "dept": "HR"}
]

def group_by_department(employees):
    departments = {}
    for i in employees:
        if i["dept"] not in departments:
            departments[i["dept"]] = []
        departments[i["dept"]].append(i["name"])
    return departments

print(group_by_department(staff))

result = group_by_department(staff)
assert "IT" in result
assert sorted(result["IT"]) == ["Алексей", "Иван"]
assert len(result["HR"]) == 2
assert result["Design"] == ["Елена"]
```

```python
staff = [
    {"name": "Алексей", "dept": "IT"},
    {"name": "Мария", "dept": "HR"},
    {"name": "Иван", "dept": "IT"},
    {"name": "Елена", "dept": "Design"},
    {"name": "Олег", "dept": "HR"}
]

def group_by_department(employees):
    departments = {}
    for i in employees:
        if i['dept'] not in departments:
            departments[i['dept']] = []
        departments.setdefault(i['dept'], []).append(i['name'])

    return departments

print(group_by_department(staff))

```

Представь, что у тебя есть два списка друзей: friends1 и friends2. Напиши функцию analyze_friends(f1, f2), которая возвращает кортеж из трех элементов:

Количество общих друзей.

Множество имен, которые есть только в первом списке.

Множество имен, которые есть только во втором списке.

```python
f1 = ["Аня", "Борис", "Влад"]
f2 = ["Влад", "Дима", "Аня"]

def analyze_friends(f1, f2):
    f_sum = 0
    f1_only = set()
    f2_only = set()

    for i in f1:
        if i in f2:
            f_sum +=1
        else:
            f1_only.add(i)

    for i in f2:
        if i not in f1:
            f2_only.add(i)

    return f_sum, f1_only, f2_only

print (analyze_friends(f1, f2))

common_count, only_f1, only_f2 = analyze_friends(f1, f2)

assert common_count == 2            # Аня и Влад
assert only_f1 == {"Борис"}
assert only_f2 == {"Дима"}
```
